```python
import numpy as np
from bqplot import *
from ipywidgets import Layout
import pandas as pd
```

```python
data = pd.read_csv('data-readonly/transportable_array/data_tohoku_norm_transpose.csv', header = None)
location = pd.read_csv('data-readonly/transportable_array/location.txt', sep = '\t', header = None)
```

```python
data.shape
data.head()
location.shape
location.head()
```

```python
location.columns = ['x', 'y', 'epi_x', 'epi_y']
epi = (142.369, 38.322)
location['epi_x'] = 142.369
location['epi_y'] = 38.322
```

```python
from haversine import haversine
location['distance'] = 0
for i in (range(len(location))):
    location['distance'].iloc[i] = haversine(location.iloc[i][:2], epi)
```

```python
location.head()
```

```python
#station 63 has no data
location.shape
data.shape
```

```python
# data normalization
data_min = data.min().min()
data_max = data.max().max()
data = (data - data_min) / (data_max - data_min)
```

```python
data.head()
```

```python
# transposing the data 
data = data.transpose()

# add the distance from the epicenter
data_distance = data
data_distance['distance'] = location['distance']
```

```python
data_distance
```

```python
# sorting the data
data_sorted = data_distance.sort_values(['distance'], ascending = False)
data_sorted = data_sorted.drop('distance', 1)
data_sorted.head()
```

```python
from ipywidgets import HBox, VBox, IntSlider, Play, jslink, HTML
import bqplot.pyplot as bqp
```

```python
# create time slider and vertical line
year_slider = IntSlider(min=0, max=4000, step=1, description='Second')
initialTime = 0
Vline = bqp.plot(x=[initialTime,initialTime],y=[0,438])
```

```python
# add slider event
def OnYearChange(change):
    currentTime = change['new']
    Vline.x = [currentTime,currentTime]
    
year_interact = year_slider.observe(OnYearChange, 'value')
```

```python
# create the x, y for the heatmap
x = np.arange(0, 4000)
y = np.arange(0, 438)
X, Y = np.meshgrid(x, y)
color = data_sorted
```

```python
# create the heatmap object
x_sc, y_sc, col_sc = LinearScale(min=0, max=4000), LinearScale(min=0, max=438), ColorScale(scheme='RdYlBu')
heat = HeatMap(x=x, y=y, color=color,
               scales={'x': x_sc, 'y': y_sc, 'color': col_sc})
```

```python
ax_x = Axis(scale=x_sc)
ax_y = Axis(scale=y_sc, orientation='vertical')
ax_c = ColorAxis(scale=col_sc)
fig = Figure(marks=[heat,Vline], axes=[ax_x, ax_y, ax_c],
             title='Spectrogram',
             layout=Layout(width='800px', height='500px'),
             min_aspect_ratio=1, max_aspect_ratio=1, padding_y=0)
VBox([fig,year_slider])
```
