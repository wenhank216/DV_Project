## Calculate the distance from epicenter

```python
#install haversine package
!pip install haversine
```

```python
import pandas as pd
```

```python
location = pd.read_table("location.txt", header=None)
```

```python
location.columns = ['x', 'y', 'epi_x', 'epi_y']
```

```python
# epicenter of Tohoku earthquake
# https://en.wikipedia.org/wiki/2011_Tōhoku_earthquake_and_tsunami
epi = (142.369, 38.322)
```

```python
location['epi_x'] = 142.369
location['epi_y'] = 38.322
```

```python
location.head()
```

```python
from haversine import haversine
```

```python
distance = haversine(location.iloc[0][:2], epi)
distance
```

```python
location.head()
```

```python
location['distance'] = 0
```

```python
for i in (range(len(location))):
    location['distance'].iloc[i] = haversine(location.iloc[i][:2], epi)
```

```python
location.head()
```

```python
location.tail()
```

```python
location = location.sort_values('distance')
```

```python
location
```
